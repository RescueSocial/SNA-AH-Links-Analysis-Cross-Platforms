```python
data_path = "../_Instagram - Data/"
```

```python
df_comments = pd.read_csv(data_path + "comments.csv")
```

```python
df_comments.head(1)
```

```python
urls_comments = URLS(df_comments)
```

```python
df_comments["timestamp"] = urls_comments.datetime_from_timestamp(
    timestamp_column="timestamp"
)
```

```python
df_comments.head(1)
```

```python
# 4 months data
df_comments_4_months = urls_comments.months_filteration(
    datetime_column="timestamp", start="2021-01-01 00:00:00", end="2021-05-01 00:00:00"
)

# dec data
df_comments_dec = urls_comments.months_filteration(
    datetime_column="timestamp", start="2020-12-01 00:00:00", end="2021-01-01 00:00:00"
)
```

```python
df_comments_4_months.timestamp.min(), df_comments_4_months.timestamp.max()
```

```python
df_comments_dec.timestamp.min(), df_comments_dec.timestamp.max()
```

```python
df_comments_4_months["urls"] = df_comments_4_months.message.apply(lambda x: get_urls(x))
```

```python
df_comments_4_months_urls = urls_df(df_comments_4_months)
```
