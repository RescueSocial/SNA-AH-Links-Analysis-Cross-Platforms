```python
import pandas as pd


df_tweets_4 = pd.read_csv("../../Twitter/df_tweets_4_months_urls.csv")
```

```python
df_tweets_4_more = df_tweets_4[df_tweets_4["n_urls"] > 1]
```

```python
df_tweets_4_more
```

```python
import ast
import requests
import urlexpander
```

```python
df_tweets_4_more["urls"] = df_tweets_4_more.urls.apply(lambda x: ast.literal_eval(x))
```

```python
b1 = df_tweets_4_more.iloc[:10].urls.values
```

```python
b1
```

```python
def unshorten_one(url):
    try:
        short = urlexpander.expand(url)
    except requests.exceptions.MissingSchema:
        short = urlexpander.expand("http://" + url)
    except requests.ConnectionError:
        short = "connection error"
    except requests.TooManyRedirects:
        short = "__CLIENT_ERROR__"
        
    if "__CLIENT_ERROR__" in short or "__CONNECTIONPOOL_ERROR__"  in short:
        try:
            r = requests.get(url, allow_redirects=False)
            try:
                short = r.headers["location"]
            except KeyError:
                short = "Page doesn't exist!"
            except requests.TooManyRedirects:
                short = urlexpander.expand(url)
        except requests.exceptions.MissingSchema:
            r = requests.get("http://" + url, allow_redirects=False)
            try:
                short = r.headers["location"]
            except KeyError:
                short = "Page doesn't exist!"
        except requests.ConnectionError:
            short = "connection error"
            

    return short
```

```python
batch1 = []

for i in b1:
    temp = []
    for v in i:
        temp.append(unshorten_one(v))
    batch1.append(temp)
```

```python
batch1
```

```python
data = []

for i in df_tweets_4_more.urls.values:
    temp = []
    for v in i:
        try:
            short = unshorten_one(v)
        except:
            short = "error"
        temp.append(short)

    data.append(temp)
```

```python
len(data)
```

```python
df_tweets_4_more["unshortened_urls"] = data
```
