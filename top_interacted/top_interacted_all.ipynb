```python
import pandas as pd
import os
import glob
```

```python
all_data = [i for i in os.listdir("Data") if i.split('.')[-1] == "csv"]
all_data
```

```python
dec_data = [i for i in all_data if "dec" in i]
months_data = [i for i in all_data if "4_months" in i]
```

```python
dec_data
```

```python
months_data
```

```python
dec = []

for filename in dec_data:
    df = pd.read_csv("Data/" + filename, index_col=None, header=0)
    dec.append(df)

df_dec = pd.concat(dec, axis=0, ignore_index=True)
```

```python
df_dec["platform"] = df_dec["url_id"].apply(lambda x: str(x).split("_")[0])
```

```python
df_dec["platform"] =df_dec["platform"].replace("nan", "reddit")
```

```python
df_dec.drop(columns="Unnamed: 0").sort_values("total", ascending=False).to_csv(
    "Data/_all_dec.csv", index=False
)
```

```python
df_dec.drop(columns="Unnamed: 0").sort_values("total", ascending=False)
```

```python
months_4 = []

for filename in months_data:
    df = pd.read_csv("Data/" + filename, index_col=None, header=0)
    months_4.append(df)

df_4_months = pd.concat(months_4, axis=0, ignore_index=True)
```

```python
df_4_months["platform"] = df_4_months["url_id"].apply(lambda x: str(x).split("_")[0])
```

```python
df_4_months = df_4_months.drop(columns="Unnamed: 0").sort_values("total", ascending=False)
```

```python
df_4_months["platform"] = df_4_months["platform"].replace("nan", "reddit")
```

```python
df_4_months.to_csv("Data/_all_4_months.csv", index=False)
```
