```python
change_AH_petitions.reset_index(drop=True, inplace=True)
```

```python
change_AH_petitions["n_shared"] = 0
change_AH_petitions["from_n_users"] = 0
change_AH_petitions["n_urls"] = 0
change_AH_petitions['id/url'] = "https://www.change.org/p/" + change_AH_petitions.petition__primary_target__slug
```

```python
change_AH_petitions.select_dtypes("float")
```

```python
change_AH_petitions.rename(columns={"_source__total_signature_count": "n_signatures",
                                   "_score": "score"}, inplace=True)
```

```python
change_AH_petitions[['n_likes', 'total', 'url_id', 'n_replies', 'retweet_count', 'favorite_count', 'n_views', 'n_dislikes', 'n_favorites', 'n_comments']] = None
```

```python
change_petitions_milestones = pd.concat(
    [
        pd.read_csv(change + "all_petitions_milestones_dec.csv"),
        pd.read_csv(change + "df_all_petitions_milestones_4.csv"),
    ]
)
```

```python
change_petitions_milestones["url"] = (
    "https://www.change.org/p/" + change_petitions_milestones.petition__slug
)
```

```python
change_petitions_milestones = change_petitions_milestones.rename(
    columns={"url": "shared_url", "signatureCount": "n_signatures", "description": "text"}
)
```

```python
change_petitions_milestones[['n_shared', 'from_n_users', 'users', 'n_urls']] = 0
```

```python
change_petitions_milestones[['platform']] = "change"
change_petitions_milestones[['platform']] = None
change_petitions_milestones[['id/url', 'n_likes', 'url_id', 'n_replies',
       'score', 'retweet_count', 'favorite_count',"from", 'n_views', 'n_dislikes',
       'n_favorites', 'n_comments']] = None
change_petitions_milestones["from"] = "petitions_milestones"
change_petitions_milestones["total"] = change_petitions_milestones["n_signatures"]
```
